### Primero tengo que agregarle día de campaña a cada observación.

```python
def add_camp_week_number(input_df):
    """Agarra un df y te devuelve el mismo df con una columna más que se llama "camp_day"
    camp_day es el número del día de la campaña. 1 = 01/07, 365 = 30/06
    también agrega una columna que es week_number. Lo mismo para el número de semana"""
    df = input_df.copy()  # copio para que cree un nuevo objeto y no toque el objeto original.
    
    # convierto la columna a tipo datetime
    df["fecha"] = pd.to_datetime(df["fecha"], format="%Y%m%d")
    df["week_number"] = df["fecha"].apply(lambda x: x.week + 26 if (x.week <= 26) else x.week - 26)
    
    return df

def add_camp_day_number(input_df):
    """Agarra un df y te devuelve el mismo df con una columna más que se llama "camp_day"
    camp_day es el número del día de la campaña. 1 = 01/07, 365 = 30/06"""
    df = input_df.copy()  # copio para que cree un nuevo objeto y no toque el objeto original.
    
    # convierto la columna a tipo datetime
    df["fecha"] = pd.to_datetime(df["fecha"], format="%Y%m%d")
    df["day_number"] = df["fecha"].apply(lambda x: x.day_of_year + 181 if (x.day_of_year < 181) else x.day_of_year - 181)
    
    return df
```

```python
def return_time_series(banda, geo, in_df):
    col_list = [i for i in range(1, 54)]
    time_series_df = pd.DataFrame(columns=["CAMP", "banda"] + col_list + ["CULTIVO"])
    dataframe = in_df.loc[:, ["CAMP", "fecha", "CULTIVO"] + [banda]]
    
    for camp, df in dataframe.groupby("CAMP"):
        df = add_camp_week_number(df)
        cultivo = df.loc[:,["CULTIVO"]].values[0]
        # Ahora transformo. (.geo va a ser un given)
        df = df.drop(axis=1, labels=["CAMP", "fecha", "CULTIVO"]).set_index("week_number").transpose()
        df.reset_index(inplace=True)
        df.rename(mapper={"index":"banda"}, axis=1, inplace=True)
        df["CAMP"] = [camp]
        df["CULTIVO"] = [cultivo[0]]
        df.index = [geo]
    
        time_series_df = pd.concat([time_series_df, df])
        
    return time_series_df
```

```python
def return_time_series(banda, geo, in_df):
    col_list = [i for i in range(1, 54)]
    time_series_df = pd.DataFrame(columns=["lote", "CAMP", "banda"] + col_list + ["CULTIVO"])
    dataframe = in_df.loc[:, ["CAMP", "fecha", "CULTIVO"] + [banda]]
    
    for camp, df in dataframe.groupby("CAMP"):
        df = add_camp_week_number(df)
        cultivo = df.loc[:,["CULTIVO"]].values[0]
        # Ahora transformo. (.geo va a ser un given)
        df = df.drop(axis=1, labels=["CAMP", "fecha", "CULTIVO"]).set_index("week_number").transpose()
        # POSIBILIDAD DE MEJORA! QUE NO DROPPEE LOS DUPLICADOS, QUE LOS PROMEDIE
        df = df.loc[:,~df.columns.duplicated()]
        # Esto está muy bueno! un tool más en la toolbox: Máscara booleana para las columnas y ~ como complemento!
        # FIN 
        df.reset_index(inplace=True)
        df.rename(mapper={"index":"banda"}, axis=1, inplace=True)
        df["CAMP"] = [camp]
        df["CULTIVO"] = [cultivo[0]]
        df["lote"] = [geo]
        try:
            time_series_df = pd.concat([time_series_df, df], ignore_index=True)
        except Exception as e:
            print(time_series_df)
            print(df)
            print(e)
            raise Exception(e)
        
    return time_series_df


def return_ts_df(input_df, bands):
    """input: table df + bands to track"""
    
    col_list = [i for i in range(1, 54)]
    time_series_df = pd.DataFrame(columns=["lote", "CAMP", "banda"] + col_list + ["CULTIVO"])
    geo_df = input_df.copy()
    
    for geo, df in geo_df.groupby(".geo"):
        for band in band_list:
            ts = return_time_series(banda=band, geo=geo, in_df=df)
            #print(ts.index)
            time_series_df = pd.concat([time_series_df, ts], ignore_index=True)
            #print(time_series_df.index)
            #print("\n")
            
    return time_series_df
```
